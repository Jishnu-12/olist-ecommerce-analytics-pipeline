#### Notebook 01: Exploratory Data Analysis

```python
import pandas as pd
import os 

# Defining Data Path using OS module for portability
DATA_PATH = os.path.join("..","Data","raw")

# Load Datasets
customers = pd.read_csv(os.path.join(DATA_PATH, "olist_customers_dataset.csv"))
geolocation = pd.read_csv(os.path.join(DATA_PATH, "olist_geolocation_dataset.csv"))
order_items = pd.read_csv(os.path.join(DATA_PATH, "olist_order_items_dataset.csv"))
payments = pd.read_csv(os.path.join(DATA_PATH, "olist_order_payments_dataset.csv"))
reviews = pd.read_csv(os.path.join(DATA_PATH, "olist_order_reviews_dataset.csv"))
orders = pd.read_csv(os.path.join(DATA_PATH, "olist_orders_dataset.csv"))
products = pd.read_csv(os.path.join(DATA_PATH, "olist_products_dataset.csv"))
sellers = pd.read_csv(os.path.join(DATA_PATH, "olist_sellers_dataset.csv"))
category_name_translation = pd.read_csv(os.path.join(DATA_PATH, "product_category_name_translation.csv"))
```

```python
# Quick overview
datasets = {
    "customers":customers,
    "geolocation":geolocation,
    "order_items":order_items,
    "payments":payments,
    "reviews":reviews,
    "orders":orders,
    "products":products,
    "sellers":sellers,
    "category_name_translation":category_name_translation
}

for name, df in datasets.items():
    print(f"{name:25} : {df.shape[0]:,} rows x {df.shape[1]} columns")
```

```python
# Detailed overview 
for name, df in datasets.items():
    print(f"{name.upper()} DATASET")
    print("\nData types and Info")
    print("-" * 20)
    df.info()
    display(df.head(3))
    print(f"Total_missing_value_count : {df.isnull().sum().sum()}")
    print(f"Total_duplicate_rows : {df.duplicated().sum():,}")
    print("-" * 145,"\n")
```

```python
# statistical analysis 
for name, df in datasets.items():
    print(f"{name.upper()} DATASET")
    display(df.describe())
    print("-" * 145,"\n")
    
```

```python
# Save summery report 
summery = {
    "Dataset":[],
    "Rows":[],
    "Columns":[],
    "Total_Missing_values":[],
    "Total_duplicate_rows":[]
}
for name,df in datasets.items():
    summery["Dataset"].append(name)
    summery["Rows"].append(f"{df.shape[0]:,}")
    summery["Columns"].append(df.shape[1])
    summery["Total_Missing_values"].append(f"{df.isnull().sum().sum():,}")
    summery["Total_duplicate_rows"].append(f"{df.duplicated().sum():,}")

summery_df = pd.DataFrame(summery)
summery_df.to_csv(os.path.join("..","Reports","Summery_report.csv"),index = False)
summery_df
```

```python
# Save Missing value count per dataset 
all_missing = {
    "Dataset":[],
    "Column_names":[],
    "Missing_value_count":[],
    "Missing_count_percentage (%)":[]
}

for name,df in datasets.items():
    missing = df.isnull().sum()
    percent = (missing / len(df)) * 100

    for col in df.columns:
       if missing[col] > 0:
           all_missing["Dataset"].append(name)
           all_missing["Column_names"].append(col)
           all_missing["Missing_value_count"].append(missing[col])
           all_missing["Missing_count_percentage (%)"].append(percent[col])
        

missing_df = pd.DataFrame(all_missing)
missing_df.to_csv(os.path.join("..","Reports","Missing_value_count_report.csv"),index = False)
missing_df
```
