#### Notebook 02: Cleaning & Merging

```python
import pandas as pd 
import numpy as np
import os

path = os.path.join("..","Data","raw")

customers = pd.read_csv(os.path.join(path,"olist_customers_dataset.csv"))
geolocation = pd.read_csv(os.path.join(path,"olist_geolocation_dataset.csv"))
order_items = pd.read_csv(os.path.join(path,"olist_order_items_dataset.csv"))
payments = pd.read_csv(os.path.join(path,"olist_order_payments_dataset.csv"))
reviews = pd.read_csv(os.path.join(path,"olist_order_reviews_dataset.csv"))
orders = pd.read_csv(os.path.join(path,"olist_orders_dataset.csv"))
products = pd.read_csv(os.path.join(path,"olist_products_dataset.csv"))
sellers = pd.read_csv(os.path.join(path,"olist_sellers_dataset.csv"))
category_name_translation = pd.read_csv(os.path.join(path,"product_category_name_translation.csv"))
```

## Customers

```python
display(customers.head(3))
print(customers.shape)
print(f"Duplicate rows: {customers.duplicated().sum()}")
print(customers.isnull().sum())
print(customers.info())
```

```python
customers.customer_city = customers.customer_city.str.title() # standardize city names for better analysis
```

```python
customers.customer_zip_code_prefix = customers.customer_zip_code_prefix.astype("str") # convert zipcode to string for treat it as an identifier
# Zip code must be string to avoid numerical operations
```

```python
customers.info()
```

## geolocation

```python
display(geolocation.head(3))
print(f"duplicate rows: {geolocation.duplicated().sum()}")
print(geolocation.isnull().sum())
print(geolocation.info())
```

```python
geolocation.geolocation_city = geolocation.geolocation_city.str.title() # standardize geolocation city
geolocation.geolocation_zip_code_prefix = geolocation.geolocation_zip_code_prefix.astype("str")
geolocation.drop_duplicates(inplace = True) # drops duplicates
```

```python
print(f"duplicated rows: {geolocation.duplicated().sum()}")
```

## order_items

```python
display(order_items.head(3))
print(order_items.info())
print(order_items.isnull().sum())
```

```python
order_items.shipping_limit_date = pd.to_datetime(order_items.shipping_limit_date) # convert datatype object to datetime
```

```python
order_items.rename(columns = {"freight_value":"freight_cost","price":"product_price"}, inplace = True) # renaming for better redability
```

```python
order_items.info()
```

## payments

```python
display(payments.head(3))
print(payments.isnull().sum())
print(f"Duplicated rows: {payments.duplicated().sum()}")
print(payments.info())
```

```python
payments.payment_type.unique()
```

```python
payments.payment_type = payments.payment_type.replace("not_defined","unknown") # replace not defined with unknown 
```

```python
payments.payment_type.unique()
```

```python
payments.rename(columns = {"payment_value":"payment_amount","payment_sequential":"payment_order"}, inplace = True) # renaming column name for redability
```

```python
payments.head()
```

## reviews

```python
display(reviews.head(3))
print(reviews.isnull().sum())
print(f"Duplicated rows: {reviews.duplicated().sum()}")
print(reviews.info())
```

```python
reviews.drop(columns = ["review_comment_title","review_comment_message"], inplace = True) # too much missing values and review in portugese
```

```python
reviews.rename(columns = {"review_score":"customer_rating"}, inplace = True) # for better redability
```

```python
# changing datatype
reviews.review_creation_date = pd.to_datetime(reviews.review_creation_date)
reviews.review_answer_timestamp = pd.to_datetime(reviews.review_answer_timestamp)
```

```python
reviews.info()
```

## orders

```python
display(orders.head(3))
print(orders.isnull().sum())
print(f"Duplicated rows: {orders.duplicated().sum()}")
print(orders.info())
```

```python
col = ["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"]

for i in col:
    orders[i] = pd.to_datetime(orders[i], errors = 'coerce') # changing datatype of every date columns

orders.info()
```

```python
orders.order_status.unique() # check unique status
```

```python
# Check order status for rows with missing delivery date
orders[orders["order_delivered_customer_date"].isnull()]["order_status"].value_counts()
```

```python
 # filtering missleading missing values delivered status containg  missing delivered date
orders[(orders.order_status == "delivered") & (orders.order_delivered_customer_date.isnull())]
```

```python
bad_row = orders[(orders.order_status == "delivered") & (orders.order_delivered_customer_date.isnull())]
orders.drop(index = bad_row.index, inplace = True) # these rows can afect performance analysis so dropped it
```

```python
orders.info()
```

## products

```python
display(products.head(3))
print(products.isnull().sum())
print(f"Duplicated rows: {products.duplicated().sum()}")
print(products.info())
```

```python
category_name_translation.head() # translation column
```

```python
products.drop(columns =
              ["product_name_lenght","product_description_lenght",
               "product_photos_qty","product_weight_g","product_length_cm",
               "product_height_cm","product_width_cm"], inplace = True) # do not need these columns for analysis
```

```python
products.head()
```

```python
products = products.merge(category_name_translation, on = "product_category_name", how = "left")
```

```python
products.isnull().sum()
```

```python
# finding difrences between two columns for finding the missing translated category name
unique1 = set(products.product_category_name) 
unique2 = set(category_name_translation.product_category_name)
missing = unique1 - unique2
missing
```

```python
products.product_category_name = products.product_category_name.replace({
    "pc_gamer":"pc_gamer","portateis_cozinha_e_preparadores_de_alimentos":"portable_kitchen_and_food_preparators"
})

products.fillna({"product_category_name":"unknown"}, inplace = True)
```

```python
products.product_category_name.unique()
```

```python
products.fillna({"product_category_name_english":"product_category_name"}, inplace=True)
```

```python
products.product_category_name = products.product_category_name_english
```

```python
products.drop(columns=["product_category_name_english"], inplace=True)
```

```python
products.head()
```

## sellers

```python
display(sellers.head(3))
print(sellers.isnull().sum())
print(f"Duplicated rows: {sellers.duplicated().sum()}")
print(sellers.info())
```

```python
sellers.seller_city = sellers.seller_city.str.title()
```

```python
# saving all cleaned files

clean_path = os.path.join("..","Data","cleaned")

customers.to_csv(os.path.join(clean_path,"customers_clean.csv"), index = False)
geolocation.to_csv(os.path.join(clean_path,"geolocation_clean.csv"), index = False)
order_items.to_csv(os.path.join(clean_path,"order_items_clean.csv"), index = False)
payments.to_csv(os.path.join(clean_path,"payments_clean.csv"), index = False)
reviews.to_csv(os.path.join(clean_path,"reviews_clean.csv"), index = False)
orders.to_csv(os.path.join(clean_path,"orders_clean.csv"), index = False)
products.to_csv(os.path.join(clean_path,"products_clean.csv"), index = False)
sellers.to_csv(os.path.join(clean_path,"sellers_clean.csv"), index = False)
```
